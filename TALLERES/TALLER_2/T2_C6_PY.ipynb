```python
# Ejercicio 1: Rutas de monitoreo (tuplas y listas)

# 1. Definir los puntos como tuplas estáticas
punto_1 = (11.1198, -74.0321)
punto_2 = (11.1250, -74.0280)
punto_3 = (11.1302, -74.0215)

# 2. Crear la ruta inicial como lista mutable
ruta_avistamiento = [punto_1, punto_2, punto_3]

# 3. Agregar el cuarto punto al final
punto_4 = (11.1355, -74.0150)
ruta_avistamiento.append(punto_4)

# 4. Extraer el último punto reportado
ultimo_reporte = ruta_avistamiento[-1]

# 5. Crear la lista paralela de elevaciones
elevaciones = [850.5, 920.0, 1050.2, 1180.8]

# 6. Calcular elevación máxima y número total de puntos
elevacion_maxima = max(elevaciones)
total_puntos = len(ruta_avistamiento)

# 7. Imprimir resultados
print("Ruta de avistamiento:", ruta_avistamiento)
print("Último reporte:", ultimo_reporte)
print("Elevaciones:", elevaciones)
print(f"Elevación máxima alcanzada: {elevacion_maxima} m")
print(f"Número total de puntos visitados: {total_puntos}")
```

```python
# Ejercicio 2: Metadatos y ecosistemas (diccionarios y conjuntos)

# 1. Crear el diccionario del parque
pnn_tayrona = {
    "nombre": "PNN Tayrona",
    "area_hectareas": 15000,
    "abierto_turismo": True,
    "ecosistemas": ["Manglar", "Bosque Seco", "Coral"]
}

# 2. Acceso seguro a la clave "fecha_creacion"
fecha_creacion = pnn_tayrona.get("fecha_creacion", "Dato no disponible")
print("Fecha de creación:", fecha_creacion)

# 3. Añadir nueva clave "departamento"
pnn_tayrona["departamento"] = "Magdalena"

# 4. Convertir observaciones crudas en conjunto para eliminar duplicados
observaciones_crudas = ["Manglar", "Coral", "Bosque Seco", "Manglar", "Coral", "Matorral"]
observaciones_unicas = set(observaciones_crudas)

# 5. Intersección entre ecosistemas oficiales y observaciones únicas
ecosistemas_oficiales = set(pnn_tayrona["ecosistemas"])
ecosistemas_avistados = ecosistemas_oficiales.intersection(observaciones_unicas)

# 6. Imprimir resultados
print("\nDiccionario actualizado del parque:")
print(pnn_tayrona)

print("\nObservaciones únicas:")
print(observaciones_unicas)

print("\nEcosistemas oficiales avistados hoy:")
print(ecosistemas_avistados)
```
