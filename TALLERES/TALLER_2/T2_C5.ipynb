```python
# Definición de variables en snake_case
nombre_estacion = "Páramo de Santurbán"
latitud = 7.2514
longitud = -72.9069
elevacion_metros = 3350
esta_activa = True

# Colección con latitud y longitud, en ese orden
coordenadas = [latitud, longitud]

# Diccionario con los metadatos de la estación
metadatos_estacion = {
    "nombre_estacion": nombre_estacion,
    "latitud": latitud,
    "longitud": longitud,
    "elevacion_metros": elevacion_metros,
    "esta_activa": esta_activa,
    "coordenadas": coordenadas
}

# Extraer la longitud desde la colección de coordenadas dentro del diccionario
longitud_extraida = metadatos_estacion["coordenadas"][1]

# Actualizar la elevación con el incremento de 12.5 metros
elevacion_metros += 12.5

# Imprimir mensaje dinámico
print(f"La estación {nombre_estacion} se encuentra operativa en la longitud {longitud_extraida} a una altura actualizada de {elevacion_metros} msnm.")
```

```python
registro_crudo = "   sAnTuaRio de fAuna Y flora iGuaQue   "

# 1) Eliminar espacios al inicio y al final
registro_limpio = registro_crudo.strip()

# 2) Convertir a Title Case
registro_limpio = registro_limpio.title()

# 3) Reemplazar " Y " por " y "
registro_limpio = registro_limpio.replace(" Y ", " y ")

# 4) Imprimir reporte con caracteres de escape
print(f"Reporte de Limpieza Toponímica:\n\tRegistro Original: \"{registro_crudo}\"\n\tRegistro Limpio: \"{registro_limpio}\"\nEstado: Listo para cruce espacial.")
```
