```python
# Ejercicio 1: Limpieza de topónimos (PNN Macarena)

registro_crudo = "   pnn  seRRania de la mACaRena   "

# Quitar espacios al inicio y al final
texto_sin_espacios = registro_crudo.strip()

# Convertir a formato título
texto_titulo = texto_sin_espacios.title()

# Reemplazar Pnn por PNN
registro_oficial = texto_titulo.replace("Pnn", "PNN")

# Calcular longitud
longitud_registro = len(registro_oficial)

# Imprimir reporte final
print(f"El área protegida limpia es: {registro_oficial} y su nombre tiene {longitud_registro} caracteres.")
```

```python
# Ejercicio 2: Parseo de coordenadas y WKT (Volcán Galeras)

mensaje_sensor = "ALERTA_SISMICA:1.221,-77.359:PROFUNDIDAD_5KM"

# Separar por :
partes = mensaje_sensor.split(":")
bloque_coords = partes[1]

# Separar por coma
lat_str, lon_str = bloque_coords.split(",")

# Convertir a float
latitud = float(lat_str)
longitud = float(lon_str)

# Construir WKT con 4 decimales
punto_wkt = f"POINT({longitud:.4f} {latitud:.4f})"

# Formatear área con separador de miles
area_afectada = 1250000
area_formateada = f"{area_afectada:,}".replace(",", ".")

# Imprimir resultados
print(f"WKT generado: {punto_wkt}")
print(f"Área afectada: {area_formateada} m²")
```
